# VPUFS


```python
import numpy as np
import pandas as pd

def calculate_variance_score(feature):
    """Calculate the variance score of a feature."""
    return np.var(feature)

def calculate_pearson_similarity_matrix(features):
    """Calculate the Pearson similarity matrix."""
    return np.corrcoef(features.T)

def calculate_non_redundant_score(pearson_sim_matrix, i):
    """Calculate the non-redundant score for a feature A_i."""
    n = pearson_sim_matrix.shape[0]
    return max(1 - pearson_sim_matrix[i, j] for j in range(n) if j != i)

def calculate_final_feature_score(variance_score, non_redundant_score):
    """Calculate the final feature score FS(A_i)."""
    return variance_score * non_redundant_score

def VPUFS(features, q):
    """Proposed feature selection method (PTUFR)."""
    n = features.shape[1]
    variance_scores = [calculate_variance_score(features.iloc[:, i]) for i in range(n)]

    pearson_sim_matrix = calculate_pearson_similarity_matrix(features)

    feature_scores = []
    for i in range(n):
        non_redundant_score = calculate_non_redundant_score(pearson_sim_matrix, i)
        final_feature_score = calculate_final_feature_score(variance_scores[i], non_redundant_score)
        feature_scores.append((features.columns[i], final_feature_score))

    # Sort features in descending order according to their scores
    feature_scores.sort(key=lambda x: x[1], reverse=True)

    # Select top q features to comprise the feature subset S
    selected_features = [feature_scores[i][0] for i in range(q)]
    return selected_features

nearest_features = VPUFS(features, 371)
print("Selected features:", nearest_features)
```

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.neighbors import NearestNeighbors
from minisom import MiniSom
from sklearn.metrics import adjusted_rand_score, rand_score
from sklearn.cluster import KMeans, AgglomerativeClustering, SpectralClustering
from sklearn.mixture import GaussianMixture

# Load your dataset from a CSV file

#X = dataset.drop('target_column', axis=1).values
#y_true = dataset['target_column'].values
X= p.iloc[:, :-1]
y_true= p.iloc[:, -1]

# k-means
kmeans = KMeans(n_clusters=6, random_state=0).fit(X)
labels_kmeans = kmeans.labels_
kmeans_rand_index = rand_score(y_true, labels_kmeans)

# Single Linkage
single_linkage = AgglomerativeClustering(n_clusters=6, linkage='single').fit(X)
labels_single_linkage = single_linkage.labels_
single_linkage_rand_index = rand_score(y_true, labels_single_linkage)

# Agglomerative
agglomerative = AgglomerativeClustering(n_clusters=6, linkage='ward').fit(X)
labels_agglomerative = agglomerative.labels_
agglomerative_rand_index = rand_score(y_true, labels_agglomerative)

# DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
y_pred_dbscan = dbscan.fit_predict(X)
dbscan_rand_index = rand_score(y_true, y_pred_dbscan)

# Spectral Clustering
spectral = SpectralClustering(n_clusters=6, random_state=42, affinity='nearest_neighbors')
y_pred_spectral = spectral.fit_predict(X)
spectral_rand_index = rand_score(y_true, y_pred_spectral)

# Gaussian Mixture Model (GMM)
gmm = GaussianMixture(n_components=6, random_state=42)
y_pred_gmm = gmm.fit_predict(X)
gmm_rand_index = rand_score(y_true, y_pred_gmm)

# SOM
n_features = p.shape[1]
som = MiniSom(4, 4, n_features, sigma=0.5, learning_rate=0.5)
som.train_random(p.values, 100)
labels_som = []
for x in p.values:
    winner = som.winner(x)
    labels_som.append(winner[0] * 4 + winner[1])
labels_som = np.array(labels_som)
som_rand_index = rand_score(y_true, labels_som)

# Print Rand Index for each clustering algorithm
print("K-Means Rand Index:", kmeans_rand_index)
print("Single Linkage Rand Index:", single_linkage_rand_index)
print("Agglomerative Rand Index:", agglomerative_rand_index)
print(f'DBSCAN Rand Index: {dbscan_rand_index}')
print(f'Spectral Clustering Rand Index: {spectral_rand_index}')
print(f'GMM Rand Index: {gmm_rand_index}')
print("SOM Rand Index:", som_rand_index)
```
